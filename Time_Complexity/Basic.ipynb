```python
def binary_search(arr, target):
    left, right = 0, len(arr)
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1



arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target = 23
binary_search(arr, target)
```

```python
# Problem #2: Find First and Last Position of an Element
# The Setup:
# You are given a sorted array of integers that may contain duplicates.

# Your Goal:
# Find the starting and ending position of a given target value. If the target is not found in the array, return [-1, -1].

# Example:

# Array: [5, 7, 7, 8, 8, 8, 10]
# Target: 8

# Expected Output: [3, 5] (since 8 starts at index 3 and ends at index 5)



nums =  [5, 7, 7, 8, 8, 8, 10]
target = 8


ans = [-1, -1]
# freq = {}
for i in range(0, len(nums)):
	if nums[i] == target and ans[0] == -1:
		ans[0] = i
	elif nums[i] == target:
		ans[1] = i
print(ans)
```

```python
# Prefix sum:


def prefix_sum(arr):
    n = len(arr)
    if n == 0:
        return []
    pref = [0]*n
    pref[0] = arr[0]
    for i in range(1, n):
        pref[i] = pref[i - 1] + arr[i]
    return pref

arr = [1,2,3,4,5]
print(arr)
print(prefix_sum(arr))
```

```python
def prefix_sum(nums):
    total_sum = 0
    for i in range(0, len(nums)):
        total_sum +=nums[i]
    left_sum = 0
    for i in range(len(nums)):
        right_sum = total_sum - left_sum - nums[i]
        if right_sum == left_sum:
            return i
        left_sum +=nums[i]

nums = [2,3,-1,8,4]
print(prefix_sum(nums))
```

```python
# Leetcode Question no. 1732

def Largest_altitude(gain):
    start_altitude = 0
    max_altitude = 0
    for i in range(len(gain)):
        start_altitude = start_altitude + gain[i]
        max_altitude = max(max_altitude, start_altitude)
    return max_altitude

gain = [-4, -3, -2, -1, 4, 3, 2]
print(Largest_altitude(gain))
```

```python
def anagram(s, t):
    if len(s) != len(t):
        return False
    
    freq = {}
    for i in s:
        freq[i] = freq.get(i, 0) + 1

    freq2 = {}
    for j in t:
        freq2[j] = freq2.get(j, 0) + 1
    if freq == freq2:
        return True
    
    return False

s = "anagram"
t = "nagaram"
print(anagram(s, t))
```

```python
def anagram(s, t):
    if len(s) != len(t):
        return False
    for i in s:
        if s.count(i) == t.count(i):
            return True
    return False
    
s = "anagram"
t = "nagaram"
print(anagram(s, t))
    
```

```python
def majority(nums):    
    freq = {}
    n = len(nums)
    for i in nums:
        freq[i] = freq.get(i, 0) + 1
        if freq[i] > n//2:
            return i
    
nums = [2,2,3,3,3,2,2]
print(majority(nums))
```

```python
def length(s):
    list = []
    S = s.strip().split()
    for i in S:
        list.append(i)
    L = list[-1]
    length_last  = len(L)
    return length_last

s = "luffy is still joyboy"
print(length(s))
```

```python
def unique(arr):
    ans_dict = {}
    for i in range(len(arr)):
        if arr[i] not in ans_dict:
            ans_dict[arr[i]] = 1
        else:
            ans_dict[arr[i]] +=1
        
    ans_list = ans_dict.values()
    if len(ans_list) == len(set(ans_list)):
        return True
    return False

arr = [2,2]
print(unique(arr))
```
