```python
def twoSum(nums, k):
    left, right = 0, len(nums)-1
    while left < right:
        sum = nums[left] + nums[right]
        if sum == k:
            return [left+1, right+1]
        elif sum > k:
            right -= 1
        else:
            left += 1

nums = [2,3,4]
k = 6
twoSum(nums, k)
```

```python
# Remove Duplicates from Sorted Array

def duplicates(nums):
    if not nums:
        return 0
    slow = 1
    for i in range(1, len(nums)):
        if nums[i] != nums[i-1]:
            nums[slow] = nums[i]
            slow += 1
    return slow

nums = [0,0,1,1,1,2,2,3,3,4]
duplicates(nums)
```

```python
# Bubble Sort

def Bubble(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped: 
            break
    return arr

arr = [5,3,8,4,2]
Bubble(arr)
```

```python
def insert_position(arr, k):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == k:
            return mid
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid - 1
    return left

arr = [1,3,5,6]
k = 7
insert_position(arr, k)
```

```python
# Build the stack using Linked List 

class Node:
    def __init__(self, val = 0, next = None):
        self.data = val
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
        self.size = 0

    def push(self, val):
        newNode = Node(val)
        newNode.next = self.head
        self.head = newNode
        self.size += 1

    def pop(self):
        if self.head == None:
            print('The stack is Empty')
            return
        last = self.head.data
        self.head = self.head.next
        self.size -= 1
        return last
    
    def peek(self):
        return self.head.data
    
    def isEmpty(self):
        if self.size == 0:
            return True
        return False
    
    def get_size(self):
        return self.size

    def traverse(self):
        curr = self.head
        while curr:
            print(curr.data, end=' -> ')
            curr = curr.next

a = Stack()
a.push(10)
a.push(20)
a.push(30)
a.push(40)
print('Peek Element :', a.peek())
print('Pop Element :', a.pop())
print('Is Empty :', a.isEmpty())
print('Size :', a.get_size())
a.traverse()
```
