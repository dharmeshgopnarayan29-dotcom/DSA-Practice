```python
# Running Sum of 1D Array

nums = [1,2,3,4]
n = len(nums)
running = [0]*n
running[0] = nums[0]
for i in range(1, n):
    running[i] = running[i-1] + nums[i]
print(running)
```

```python
# Subarray Sum Equals K

nums = [1,1,1]
k = 2

sub_num = {0:1}
total = count = 0
for n in nums:
    total += n
    if total - k in sub_num:
        count += sub_num[total-k]
    sub_num[total] = 1 + sub_num.get(total, 0)
print(count)
```

```python
def twoSum(nums, k):
    left, right = 0, len(nums)-1
    while left < right:
        sum = nums[left] + nums[right]
        if sum == k:
            return [left+1, right+1]
        elif sum > k:
            right -= 1
        else:
            left += 1

nums = [2,3,4]
k = 6
twoSum(nums, k)
```
