```python
# Example 1: Find All Duplicates in an Array

# Input: nums = [4,3,2,7,8,2,3,1]
# Output: [2,3]

def Duplicate(nums):
    already = []
    dup = []
    if len(nums) == 1:
        return dup 
    for i in nums:
        if i not in already:
            already.append(i)
        else:
            dup.append(i)
    return dup

nums = [4,3,2,7,8,2,3,1]
print("Nummber 1", Duplicate(nums))

nums2 = [2,3,4,5,5,4,5,4,5,6,7]       # In This, Duplicate repeating in final result [5, 4, 5, 4, 5]
print("Nummber 2", Duplicate(nums2))

nums3 = [1]
print("Nummber 3", Duplicate(nums3))


# Second Method

def Duplicate1(nums):
    count = {}                # This dictionary will store {number: count}
    dupl = []
    for num in nums:
        current = count.get(num, 0)
        count[num] = current + 1
        if count[num] == 2:
            dupl.append(num)
    return dupl

nums4 = [4,3,2,7,8,2,3,1]
print("Nummber 4", Duplicate1(nums4))

nums5 = [2,3,4,5,5,4,5,4,5,6,7]
print("Nummber 5", Duplicate1(nums5))

nums6 = [1]
print("Nummber 6", Duplicate1(nums6))

# Third Method

def Duplicate2(nums):
    nums_copy = list(nums)
    nums_copy.sort()
    dupl = []
    if len(nums_copy) <= 1:
        return []
    for i in range(1, len(nums_copy)):
        if nums_copy[i] == nums_copy[i-1]:
            if not dupl or dupl[-1] != nums_copy[i]:
                dupl.append(nums_copy[i])
    return dupl 

nums7 = [4,3,2,7,8,2,3,1]
print("Nummber 7", Duplicate2(nums7))

nums8 = [2,3,4,5,5,4,5,4,5,6,7]
print("Nummber 8", Duplicate2(nums8))

nums9 = [1]
print("Nummber 9", Duplicate2(nums9))
```
