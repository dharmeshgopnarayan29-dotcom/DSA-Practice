```python
# Efficient Implementation: Build an Queue using Deque

# Deque (Double-ended Queue) is an efficient way to implement a queue because 
# it allows adding and removing from both end in O(1) time.

from collections import deque
que = deque()

que.append(1)
que.append(2)
que.append(3)
que.append(4)
que.appendleft(5)
print(que)

que.pop()
que.popleft()
print(que)
```

```python
#  Implement Fixed Size Queue (Size N) using Arrays

class Queue:
    def __init__(self, n):
        self.queue = [None]*n
        self.front = -1
        self.rear = -1
        self.size = n

    def enqueue(self, val):
        if self.rear == self.size - 1:
            print("Queue Overflow")
            return
        
        if self.front == -1:
            self.front = 0

        self.rear += 1
        self.queue[self.rear] = val

    def dequeue(self):
        if self.front == -1 or self.front > self.rear:
            print("Queue OverFlow")
            return
        
        value = self.queue[self.front]
        self.front += 1
        return value
    
    def display(self):
        if self.front == -1:
            print("Queue is empty")
        else:
            for i in range(self.front, self.rear+1):
                print(self.queue[i], end=" ")
            print()
        
q = Queue(5)

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.display()

q.dequeue()
q.display()
```

```python
# Sliding Window Maximum Problem

# Find maximum element in every window of size k.

from collections import deque

def max_window(arr, k):
    dq = deque()
    res = []

    for i in range(len(arr)):
        while dq and dq[0] <= arr[i]:
            dq.popleft()

        while dq and arr[dq[-1]] < arr[i]:
            dq.pop()

        dq.append(i)

        if i >= k-1:
            res.append(arr[dq[0]])

    return res

arr = [1,3,-1,-3,5,3,6,7]
k = 3
max_window(arr, k)
```

```python
# Implement Stack using Queue

from collections import deque

class Stack:
    def __init__(self):
        self.q1=deque()
        self.q2=deque()

    def push(self,x):
        self.q2.append(x)

        while self.q1:
            self.q2.append(self.q1.popleft())

        self.q1,self.q2=self.q2,self.q1

    def pop(self):
        return self.q1.popleft()
```

```python
# Implement Queue using Stack

class Queue:
    def __init__(self):
        self.s1=[]
        self.s2=[]

    def enqueue(self,x):
        self.s1.append(x)

    def dequeue(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2.pop()
```
