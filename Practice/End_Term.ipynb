```python
def binary(arr, k):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == k:
            return mid
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid -1
    return -1

arr = [1,2,3,4,5,5,7,8]
binary(arr, 5)
```

```python
def first_occ(arr, target):
    left, right = 0, len(arr) - 1
    ans = -1

    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            right = mid - 1
            ans = mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return ans

arr = [1,2,3,4,5,5,7,8]
first_occ(arr, 5)
```

```python
# Recursive (Binary search)

def binary_recursive(arr, k, left, right):
    while left > right:
        return -1
    mid = (left + right)//2
    if arr[mid] == k:
        return mid
    elif arr[mid] < k:
        return binary_recursive(arr, k, mid+1, right)
    else:
        return binary_recursive(arr, k, left, mid-1)
    
arr = [1,2,3,4,5,6,7,8]
binary_recursive(arr, 7, 1, 8)
```

```python
# Reverse an array

def reverse(arr):
    left, right = 0, len(arr)-1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr

arr = [1,2,3,4,5,7,8]
reverse(arr)
```

```python
def prefix_sum(arr):
    n = len(arr)
    if n == 0:
        return []
    pref = [0]*n
    pref[0] = arr[0]
    for i in range(1,n):
        pref[i] = pref[i-1] + arr[i]
    return pref

arr = [1,2,3,4,5,6]
prefix_sum(arr)
```

```python
# 49 - Group Anagrams

def Group_anagrams(str):
    anagrams = {}
    for i in str:
        sorted_word = tuple(sorted(i))
        if sorted_word in anagrams:
            anagrams[sorted_word].append(i)
        else:
            anagrams[sorted_word] = [i]
    result = []
    for key in anagrams:
        result.append(anagrams[key])
    return result
str = ["eat","tea","tan","ate","nat","bat"]
Group_anagrams(str)
```

```python
def rotate(arr, k):
    n = len(arr)
    k = k % n
    rotated = [0] * n
    for i in range(n):
        rotated[(i + k) % n] = arr[i]
    return rotated
arr = [1,2,3,4,5,6,7]
k = 3
rotate(arr, k)
```

```python
def merge(arr1,m, arr2,n):
    left, right = 0, 0
    res = []
    while left < m and right < n:
        if arr1[left] < arr2[right]:
            res.append(arr1[left])
            left += 1
        else:
            res.append(arr2[right])
            right += 1
    while left < m:
        res.append(arr1[left])
        left += 1
    while right < n:
        res.append(arr2[right])
        right += 1
    for i in range(len(res)):
        arr1[i] = res[i]
    
    return arr1

arr1 = [1,2,3,0,0,0]
m = 3
arr2 = [2,5,6]
n = 3
merge(arr1,m, arr2,n)
```

```python
def Sum_equals_k(arr, k):
    res = []
    for i in arr:
        diff = k - i
        if diff < k and diff in arr:
            res.append(diff)
            res.append(i)
    return len(res)
arr = [1,2,3]
k = 3
Sum_equals_k(arr,k)
```

```python
def isPalindrome(s):
    left, right = 0, len(s)-1
    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if left < right:
            if s[left].lower() != s[right].lower():
                return False
            left += 1
            right -= 1
    return True
```

```python
def color_sort(arr):
    left, right = 0, len(arr)-1
    i = 0
    while i <= right:
        if arr[i] == 0:
            arr[i], arr[left] = arr[left], arr[i]
            left += 1
            i += 1
        elif arr[i] == 2:
            arr[i], arr[right] = arr[right], arr[i]
            right -= 1
        else:
            i += 1
    return arr

arr = [2, 0, 2, 1, 1, 0]
color_sort(arr)
```

```python
def Best_Time_To_sell(arr):
    max_profit = 0
    min_price = arr[0]
    for i in range(len(arr)):
        if min_price > arr[i]:
            min_price = arr[i]
        elif arr[i] - min_price > max_profit:
            max_profit = arr[i] - min_price
    return max_profit

arr = [7,1,5,3,6,4]
Best_Time_To_sell(arr)
```

```python
def isIsomorphic(s, t):
    m1 = {}
    m2 = {}
    for i in range(len(s)):
        if s[i] in m1 and m1[s[i]] != t[i]:
            return False
        if t[i] in m2 and m2[t[i]] != s[i]:
            return False
        m1[s[i]] = t[i]
        m2[t[i]] = s[i]
    return True, m1, m2

s = "egg"
t = "add"
isIsomorphic(s, t)
```

```python
def Longest_common_prefix(arr):
    if not arr:
        return ""
    
    for i in range(len(arr[0])):
        char = arr[0][i]
        for j in arr[1:]:
            if i >= len(j) or j[i] != char:
                return arr[0][:i]
    return arr[0]

arr = ["flower", "flow", "flight"]
Longest_common_prefix(arr)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
    
class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, val):
        new = Node(val)
        if self.head is None:
            self.head = new
            return
        new.next = self.head
        self.head.prev = new
        self.head = new

    def delete_position(self, k):
        if self.head is None:
            print("Nothing to delete")
            return
        if k == 1:
            self.head = self.head.next
            if self.head:
                self.head.prev = None
            return

        temp = self.head
        curr = 1
        while temp and curr < k:
            temp = temp.next 
            curr += 1
        if temp is None:
            print("Invalid position")
            return
        if temp.next:
            temp.next.prev = temp.prev
        if temp.prev:
            temp.prev.next = temp.next

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

l = LinkedList()
l.insert(10)
l.insert(20)
l.insert(30)
l.insert(40)
print("Before deletion")
l.display()
print("After deletion")
l.delete_position(4)
l.display()
```
