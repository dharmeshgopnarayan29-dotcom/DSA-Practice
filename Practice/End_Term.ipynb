```python
def binary(arr, k):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == k:
            return mid
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid -1
    return -1

arr = [1,2,3,4,5,5,7,8]
binary(arr, 5)
```

```python
def first_occ(arr, target):
    left, right = 0, len(arr) - 1
    ans = -1

    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            right = mid - 1
            ans = mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return ans

arr = [1,2,3,4,5,5,7,8]
first_occ(arr, 5)
```

```python
# Recursive (Binary search)

def binary_recursive(arr, k, left, right):
    while left > right:
        return -1
    mid = (left + right)//2
    if arr[mid] == k:
        return mid
    elif arr[mid] < k:
        return binary_recursive(arr, k, mid+1, right)
    else:
        return binary_recursive(arr, k, left, mid-1)
    
arr = [1,2,3,4,5,6,7,8]
binary_recursive(arr, 7, 1, 8)
```

```python
def binary_last(arr, k):
    left, right = 0, len(arr)-1
    ans = -1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == k:
            ans = mid
            left = mid + 1
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid - 1
    return ans
arr = [1,2,3,4,4,5,5,7,8]
binary_last(arr, 5)
```

```python
def binary_floor(arr, k):
    left, right = 0, len(arr)-1
    ans = -1
    
    while left <= right:
        mid = (left + right)//2
        if arr[mid] <= k:
            ans = arr[mid]
            left = mid + 1
        else:
            right = mid - 1
    return ans

arr = [1,2,3,4,6,7,8]
binary_floor(arr, 5)
```

```python
def binary_ceil(arr, k):
    left , right = 0, len(arr)-1
    ans = -1

    while left <= right:
        mid = (left + right)//2
        if arr[mid] >= k:
            ans = arr[mid]
            right = mid - 1
        else:
            left = mid + 1
    return ans
arr = [1,2,3,4,5,7,8]
binary_ceil(arr, 6)
```

```python
def binary_insert(arr, k):
    left, right = 0, len(arr)-1

    while left <= right:
        mid = (left + right)//2
        if arr[mid] == k:
            return mid
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid - 1
    return left

arr = [1,2,3,4,5,7,8]
binary_insert(arr, 6)
```

```python
def binary_peak(arr):
    left, right = 0, len(arr)-1

    while left < right:
        mid = (left + right)//2
        if arr[mid] > arr[mid + 1]:
            right = mid
        else:
            left = mid + 1
    return arr[left]
arr = [1,2,3,4,5,7,8]
binary_peak(arr)
```

```python
def peak_index(arr):
    low, high = 0, len(arr) - 1
    
    while low < high:
        mid = (low + high) // 2
        if arr[mid] > arr[mid + 1]:
            high = mid
        else:
            low = mid + 1
    return low
arr = [1,2,3,4,5,7,8]
peak_index(arr)
```

```python
# Reverse an array

def reverse(arr):
    left, right = 0, len(arr)-1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr

arr = [1,2,3,4,5,7,8]
reverse(arr)
```

```python
def duplicate(arr):
    a = set()
    dup = []
    for i in arr:
        if i not in a:
            a.add(i)
        else:
            dup.append(i)
    return dup
arr = [1]
duplicate(arr)
```

```python
def missing(arr):
    n = len(arr)
    total = (n*(n+1))//2
    sum = 0
    for i in range(len(arr)):
        sum += arr[i]
    return total - sum
arr = [3,0,1,4,5,6,2,8]
missing(arr)
```

```python
def prefix_sum(arr):
    n = len(arr)
    if n == 0:
        return []
    pref = [0]*n
    pref[0] = arr[0]
    for i in range(1,n):
        pref[i] = pref[i-1] + arr[i]
    return pref

arr = [1,2,3,4,5,6]
prefix_sum(arr)
```

```python
def middle_index(arr):
    total = 0
    for i in arr:
        total += i
    left = 0
    for i in range(len(arr)):
        right = total - left - arr[i]
        if left == right:
            return i
        left += arr[i]
    return -1
arr = [2,3,-1,8,4]
middle_index(arr)
```

```python
def highest_altitude(arr):
    start = 0
    max1 = 0
    for i in arr:
        start += i
        max1 = max(max1, start)
    return max1

arr = [-5,1,5,0,-7]
highest_altitude(arr)
```

```python
def valid(s,t):
    if len(s) != len(t):
        return False
    for i in set(s):
        if s.count(i) != t.count(i):
            return False
    return True
s = "anagram"
t = "nagaram"
valid(s,t)
```

```python
def Unique_char(s):
    freq = {}
    for i in s:
        freq[i] = freq.get(i, 0)+1
    for i in range(len(s)):
        if freq[s[i]] == 1:
            return i
    return -1

s = "leetcode"
Unique_char(s)
```

```python
def Unique(s):
    for i in range(len(s)):
        for j in range(i+1,len(s)):
            if s[i] == s[j]:
                return i
    return -1

s = "leetcode"
Unique_char(s) 
```

```python
def majority(arr):
    count = {}
    for i in arr:
        count[i] = count.get(i, 0)+1
        if count[i] > len(arr)//2:
            return i
    return -1
arr  = [2,3,2,3,3,3,2]
majority(arr)
```

```python
def Occurrence(s):
    freq = {}
    for i in s.lower():
        freq[i] = freq.get(i, 0) + 1
    value = list(freq.values())
    for j in value:
        if j != value[0]:
            return False
    return True
```

```python
def length_last(s):
    length = 0
    n = len(s)-1
    while n >= 0 and s[n] == " ":
        n -= 1
    while n >= 0 and s[n] != " ":
        n -= 1
    return length
```

```python
def Kadane(arr):
    current = arr[0]
    max1 = arr[0]
    for i in range(1, len(arr)):
        current = max(arr[i], current + arr[i])
        max1 = max(max1, current)
    return max1
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
Kadane(arr)
```

```python
def remove(arr):
    i = 0
    while i < len(arr)-1:
        if arr[i] == arr[i+1]:
            arr.pop(i+1)
        else:
            i += 1
    return arr
arr = [1,1,2,3,3]
remove(arr)
```

```python
def two_sum(arr, k):
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] + arr[j] == k:
                return [i,j]
arr = [1,2,3,4,5,6]
two_sum(arr,6)
```

```python
# Only work in sorted array
def two_sum(arr, k):
    left, right = 0, len(arr)-1
    while left < right:
        current = arr[left] + arr[right]
        if current == k:
            return [left+1, right+1]
        elif current < k:
            left += 1
        else:
            right -= 1
```

```python
def Second_largest(arr):
    first = second = -1
    for i in arr:
        if i > first:
            second = first
            first = i
        elif i > second and i != first:
            second = i
    return second
arr = [12, 35, 1, 10, 34, 1]
Second_largest(arr)
```

```python
def plus_one(arr):
    n = len(arr)
    for i in range(n-1, -1, -1):
        if arr[i] < 9:
            arr[i] += 1
            return arr
        else:
            arr[i] = 0
    return [1] + arr
arr = [4,3,2,1]
arr2 = [9]
plus_one(arr)
plus_one(arr2)
```
