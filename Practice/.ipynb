```python
# Search for a given number in an array using Linear Search.

def search(arr, n):
    for i in range(len(arr)):
        if arr[i] == n:
            return f"Found number: {arr[i]} at index {i}..."
    return -1

arr = [1,2,3,4,5,9,6,7]
n = 8
search(arr, n)
```

```python
# Find the index of the first occurrence of a target element.

def search(arr, k):
    for i in range(len(arr)):
        if arr[i] == k:
            return f"First occurrence of a target at index {i}"
    return f"Number not found"

arr = [1,2,3,4,1,2,5,6,7]
k = 2
search(arr, k)
```

```python
# Count how many times a given element appears in the array.

def search(arr, n):
    a = 0
    for i in range(len(arr)):
        if arr[i] == n:
            a += 1
    if n not in arr :
        return f"Element not found in arr..."   
    return a
    

arr = [1,2,3,4,3,4,4,5,4,4]
n = 9
search(arr, n)
```

```python
# Find the maximum element using Linear Search.

def search(arr):
    max_element = arr[0]
    n = len(arr)-1
    for i in range(n):
        if arr[i+1] > max_element:
            max_element = arr[i+1]
    return max_element

arr = [188,2,3,65,5,42,8,9]
search(arr)
```

```python
# Find the minimum element using Linear Search.

def search(arr):
    n = len(arr)-1
    min = arr[0]
    for i in range(n):
        if arr[i+1] < min:
            min = arr[i+1]
    return min

arr = [188,2,3,65,5,42,8,1]
search(arr)
```

```python
# Return True if a number exists in a list, else False.

def search(list1, n):
    for i in list1:
        if i == n:
            return True
    return False
        
list1 = [1,2,3,4,5,6,6,7,8]
n = 7
search(list1, n)
```

```python
# Print all indexes where the target element occurs

def search(arr, k):
    index = []
    n = len(arr)
    for i in range(n):
        if arr[i] == k:
            index.append(i)
    return index

arr = [1,2,3,4,4,4,5,4,4,5,6,4,4]
k = 4
search(arr, k)
```

```python
# Check if an array is sorted using linear scanning.

def search(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            return False
    return True

arr1 = [1,2,3,6,3,6,8]
arr2 = [1,2,3,4,5,6,7,8]
search(arr2)
```

```python
# Find the first repeated element using Linear Search.

def search(arr):
    for i in range(len(arr)):
            for j in range(i+1, len(arr)):
                if arr[j] == arr[i]:
                    return arr[i]
                
arr = [1,2,3,4,5,6,7,8,7]
search(arr)
```

```python
# Search for a number in a list of tuples (search in first item).

def tuple_search(data, k):
    for i in data:
        if i[0] == k:
            return i[0]
    return -1
        
data = [(10, "A"), (20, "B"), (30, "C"), (40, "D")]
tuple_search(data, 40)
```
