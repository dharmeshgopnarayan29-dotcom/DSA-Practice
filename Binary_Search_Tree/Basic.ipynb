```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def InOrder(root):
    if root:
        InOrder(root.left)
        print(root.data, end=" -> ")
        InOrder(root.right)

#          1
#         / \
#        2   3

root = Node(1)
root.left = Node(2)
root.right = Node(3)
InOrder(root)

```
