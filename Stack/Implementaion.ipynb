```python
# Implementation Using Python List

class Stack_list:
    def __init__(self):
        # Initialize empty stack
        self.items = []

    def push(self, item):
        # Add Item to the Top of Stack
        self.items.append(item)

    def pop(self):
        # Remove and return Top item 
        if self.is_empty():
            raise IndexError("Stack is Empty - Cannot pop")
        item = self.items.pop()
        return item
    
    def peek(self):
        # Return Top item without Removing it
        if self.is_empty():
            raise IndexError("Stack is empty - cannot peek")
        return self.items[-1]
    
    def size(self):
        # Return Number of items
        return len(self.items)
    
    def is_empty(self):
        # Check if Stack is Empty 
        if len(self.items) == 0:
            return True
        return False
    
    def display(self):
        # Display Stack From Bottom to Top
        print("Stack :", self.items)

stack = Stack_list()
stack.push(10)      
stack.push(20)      
stack.push(30)    
stack.display() 
print("Top element:", stack.peek()) 
stack.pop()  
stack.display()      
print("Size:", stack.size())  
print("Is empty?", stack.is_empty())  
```

```python
# Stack Implementation Using Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class StackLinkedList:
    def __init__(self):
        # Initialize Empty Stack
        self.top = None                # Pointer to top
        self.count = 0                 # Number of Elements

    def push(self, item):
        # Add item to Top od Stack
        new_node = Node(item)
        new_node.next = self.top          # New Node points to current top
        self.top = new_node               # Update top to new node
        self.count += 1
        self.display()

    def pop(self):
        # Remove and return Top item
        if self.is_empty():
            raise IndexError("Stack is Empty - Cannot pop")
        item = self.top.data
        self.top = self.top.next
        self.count -= 1
        self.display()
        return item
    
    def peek(self):
        # Return Top item without Removing it
        if self.is_empty():
            raise IndexError("Stack is Empty - Cannot peek")
        return self.top.data
    
    def is_empty(self):
        # Display Stack from top to bottom 
        if self.count == 0:
            return True
        return False
    
    def size(self):
        return self.count
    
    def display(self):
        if self.is_empty():
            print("Stack is Empty")
            return
        
        current = self.top
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print()


stack = StackLinkedList()
stack.push(10)
stack.push(20)
stack.push(30)
print("Top element:", stack.peek())  
print("Size:", stack.size())     
# stack.pop() 
```

```python
# Balanced Parentheses

def is_balanced(s):
    stack = []

    pairs = {')':'(', '}':'{', ']':'['}

    for ch in s:
        if ch in "({[":
            stack.append(ch)

        else:
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()

    return len(stack) == 0


print(is_balanced("{[()]}"))
```

```python
# Next Greater Element 

def next_greater(arr):
    stack = []
    result = []

    for i in range(len(arr)-1, -1, -1):
        while stack and stack[-1] <= arr[i]:
            stack.pop()

        if not stack:
            result.append(-1)
        else:
            result.append(stack[-1])
        
        stack.append(arr[i])

    result.reverse()
    return result

arr = [4,5,2,10]
next_greater(arr)
```

```python
# PostFix Evaluation Using Stack

# Algorithm :
    # 1. Scan expression from left to right
    # 2. If operand → push into stack
    # 3. If operator → pop two operands
    # 4. Apply operation
    # 5. Push result back into stack

def postFix(exp):
    stack = []

    for i in exp:
        if i.isdigit():
            stack.append(int(i))
        else:
            b = stack.pop()
            a = stack.pop()

            if i == "+":
                stack.append(a+b)
            elif i == "-":
                stack.append(a-b)
            elif i == "*":
                stack.append(a*b)
            elif i == "/":
                stack.append(a/b)

    return stack[0]

postFix("23*54*+9-")
```

```python
# PreFix Evaluation Using Stack

# Algorithm :
    # 1. Scan expression right to left
    # 2. If operand → push into stack
    # 3. If operator → pop two operands
    # 4. Apply operation
    # 5. Push result back

def evaluate_prefix(exp):
    stack = []

    for ch in reversed(exp):
        if ch.isdigit():
            stack.append(int(ch))

        else:
            a = stack.pop()
            b = stack.pop()

            if ch == '+':
                stack.append(a + b)
            elif ch == '-':
                stack.append(a - b)
            elif ch == '*':
                stack.append(a * b)
            elif ch == '/':
                stack.append(a / b)

    return stack[0]

print(evaluate_prefix("-+*23*549"))
```
